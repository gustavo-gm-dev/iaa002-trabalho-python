## b) Verifique se há valores faltantes nos dados. Caso haja, escolha uma tratativa para resolver o problema de valores faltantes

```python
# Verificando valores faltantes nos dados
dados.isna().any()
```

```python
# Todas as colunas possuem valores faltantes, vamos verificar a quantidade de valores faltantes em cada coluna
dados.isna().sum()
```

```python
# Número de linhas com todos os valores faltantes
dados[dados.isna().all(axis=1)].shape
```

Constatei que os dados faltantes são várias linhas que possuem TODAS as colunas faltantes. (Número de linhas com todos os valores faltantes é igual ao número de valores faltantes em todas as colunas)
Portanto, faz mais sentido removê-las do que tentar inferir algum dado.

```python
dados=dados.dropna() # remover linhas com valores faltantes
dados.isna().sum() # comprova que o número de valores faltantes agora é zero
```

```python
# Conferindo o número de linhas e colunas do dataframe após a limpeza
dados.shape
```

## c) Verifique se há dados duplicados nos dados

```python
# Soma a quantidade de dados duplicados (linhas que possuem os mesmos valores em todas as colunas)
dados.duplicated().sum()
```

```python
# Como são apenas 2 linhas, quero dar uma olhada nelas para decidir se quero removê-las ou não
dados[dados.duplicated()].head()
```

```python
# OK... vamos removê-las!
dados.drop_duplicates(inplace=True)
# Obs: O parâmetro inplace=True faz com que a remoção seja feita no próprio dataframe, sem precisar atribuir o resultado a uma nova variável
```

```python
# Conferindo o número de linhas e colunas do dataframe após a limpeza
dados.shape
```

No caso, reduziram de 202297 linhas para 202295, confirmando nossa remoção.

## d) Crie duas categorias, para separar colunas numéricas e categóricas. Imprima o resumo de informações das variáveis numéricas e categóricas (estatística descritiva dos dados)

```python
# Vou dar uma olhadinha na tipagem dos dados antes de quebrar o dataset
dados.dtypes
```

```python
# O engine_size está como string, mas aparentemente é numérico. Vou verificar os valores únicos dessa coluna para entender o que está acontecendo
dados['engine_size'].unique()
```

```python
# O engine_size tem valores numéricos, o que pode ser uma boa variável para o modelo (ex: quanto maior o motor, mais caro o carro).
# Vou converter o separador decimal de vírgula para ponto e depois converter a coluna para numérica
dados['engine_size'] = dados['engine_size'].str.replace(',', '.').astype(float)
```

```python
dados.dtypes # conferindo a tipagem dos dados após a conversão
```

```python
# Outro dado que também precisa de conversão é o month_of_reference que está em string.
# Vou criar uma coluna nova chamada month_of_reference_num com os meses convertidos para números (ex: janeiro = 1, fevereiro = 2, etc)
# Assim a coluna pode ser usada para o modelo, trazendo uma ordem de grandeza para os meses (ex: quanto mais avançado no ano, menor o valor do carro).

# Criando um dicionário para mapear os meses em números
meses = {
    'January': 1,
    'February': 2,
    'March': 3,
    'April': 4,
    'May': 5,
    'June': 6,
    'July': 7,
    'August': 8,
    'September': 9,
    'October': 10,
    'November': 11,
    'December': 12
}
# Criando a nova coluna month_of_reference_num usando o dicionário de meses
dados['month_of_reference_num'] = dados['month_of_reference'].map(meses)
```

```python
# Conferindo se deu certo:
dados['month_of_reference_num'].unique()
```

```python
dados.dtypes # conferindo a tipagem dos dados após a conversão
```

```python
# Criando categorias para separar numéricas e categóricas
numericas_cols = [col for col in dados.columns if dados[col].dtype != 'str']
categoricas_cols = [col for col in dados.columns if dados[col].dtype == 'str']
```
